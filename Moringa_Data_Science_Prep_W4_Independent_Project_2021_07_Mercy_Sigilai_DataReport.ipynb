```python
# import pandas
import pandas as pd

# import numpy
import numpy as np

# import datetime
from datetime import datetime
```

```python
# Reading dataset from csv file

Autolib = pd.read_csv('http://bit.ly/autolib_dataset')
Autolib.head()
```

```python
# To extract information of the dataset

Autolib.info()
```

```python
# Checking for missing values

Autolib.isnull().sum()

# Displayed comment and Scheduled at has null values and since they are not needed in the analysis, we drop them
Autolib.drop(columns=['Scheduled at', 'Displayed comment'], inplace=True)
Autolib
```

```python
# Checking for duplicates
Autolib.duplicated().sum() 

# There are no duplicates
```

```python
# Select paris from city column

Autolib_paris = Autolib[Autolib['City'] =='Paris'] 

Autolib.head()
```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

Autolib_paris.groupby('Bluecar counter')['hour'].count().sort_values(ascending= True).head(1)
```

```python
# What is the most popular hour for returning cars?

Autolib_paris.groupby('Bluecar counter')['hour'].count().sort_values(ascending= False).head(1)
```

```python
# What station is the most popular?
# Overall?
Autolib[(Autolib['Kind'] == 'STATION') & (Autolib['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending= False).head(1)
```

```python
 # What station is the most popular at the most popular picking hour?
Autolib[(Autolib['Kind'] == 'STATION') & (Autolib['Status'] == 'ok') & (Autolib['hour'] == 4)].groupby('Public name')['Public name'].count().sort_values(ascending= False).head(1)
```

```python
# What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?
# Overall?

Autolib[(Autolib['Kind'] == 'STATION') & (Autolib['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False).head(1)

```

```python
# What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code at the most popular picking hour?

Autolib[(Autolib['Kind'] == 'STATION') & (Autolib['Status'] == 'ok') & (Autolib['hour'] == 4)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False).head(1)
```
